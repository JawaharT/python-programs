```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as warn
warn.filterwarnings('ignore')
%matplotlib inline
```

```python
controls_data = pd.read_excel('Controls data.xlsx')
controls_data.head()
```

```python
sns.set_style('whitegrid')
f,axes = plt.subplots(1,3, figsize=(18,6))

lt_plot = sns.scatterplot(data=controls_data, x='Time', y='LT', s=100, ax=axes[0], edgeColor='Blue', color='Black')
flow_plot = sns.scatterplot(data=controls_data, x='Time', y='Flow', s=100, ax=axes[1], edgeColor='Blue', color='Black')
pump2_plot = sns.scatterplot(data=controls_data, x='Time', y='Pump 2', s=100, ax=axes[2], edgeColor='Blue', color='Black')

plt.show()
```

```python
controls_data_copy = controls_data.copy()
controls_data_copy['LT_acceleration'] = pd.Series(controls_data_copy['Pump 2']/controls_data_copy['Time'], index=controls_data_copy.index)

sns.set_style('darkgrid')
lt_flow_acc_plot = sns.lineplot(data=controls_data_copy, x='LT_acceleration', y='Flow')

plt.xlabel('LT Acceleration', fontsize=20, color='Black', fontname='Calibri body')
plt.ylabel('Flow', fontsize=20, color='Black', fontname='Calibri body')
plt.show()
```
